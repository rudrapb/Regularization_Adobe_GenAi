```python
# this is the code to regularise a doodle

import numpy as np
import cv2
img=cv2.imread('shape1.png')
img=cv2.resize(img, (600,500))
cv2.imshow("original", img)
#lets detect shapes
imggray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,thresh=cv2.threshold(imggray, 240,255,cv2.THRESH_BINARY) #find the threshold of color in image
contours,_=cv2.findContours(thresh, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) #IT WILL FIND THE CONTOUR IN IMAGE
for contour in contours:
    approx=cv2.approxPolyDP(contour,0.002*cv2.arcLength(contour,True),True) #this method approximate polygonal curve
    cv2.drawContours(img,[approx],0,(0,0,0),3) #to draw contour
    x=approx.ravel()[0]
    y=approx.ravel()[1]-7 #defining the x and y point of shape to write text on it
    #now if else conditions to find different shapes
    #for triangle
    if len(approx)==3: #triangle has 3 sides
        cv2.putText(img,(x,y),cv2.FONT_HERSHEY_COMPLEX,0.7,(255,255,255),2)
    #for square and rectangle
    elif len(approx)==4:
        x,y,w,h=cv2.boundingRect(approx)
        if(x+w)==(y+h): #condition for square
            cv2.putText(img,"",(x, y-5), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255, 255, 255), 2)
        else:# for rectangle
            cv2.putText(img, "",(x, y-10), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255, 255, 255), 2)
    elif len(approx)==6:
        cv2.putText(img, "", (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255, 255, 255), 2)
    elif len(approx)==5:
        cv2.putText(img, "", (x-40, y), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255, 255, 255), 2) #full polygon name is not visible
    elif len(approx)==10:
        cv2.putText(img, "", (x-35, y), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255, 255, 255), 2) #now its good
    else:
        cv2.putText(img, "", (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255, 255, 255), 2)
cv2.imshow("Shape", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# symmetry: ( for this code only one image should be there to find the line of symmetry;

```python
import cv2
import numpy as np

def find_symmetry_lines(image_path):
    # Load the image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Ensure the image is binary (black and white)
    _, binary_img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)
    
    # Invert the image for better results with symmetry detection
    binary_img = cv2.bitwise_not(binary_img)
    
    # Detect the vertical line of symmetry
    vertical_symmetry = detect_symmetry(binary_img, direction='vertical')
    
    # Detect the horizontal line of symmetry
    horizontal_symmetry = detect_symmetry(binary_img, direction='horizontal')
    
    # Display results
    print(f"Vertical line of symmetry: {vertical_symmetry}")
    print(f"Horizontal line of symmetry: {horizontal_symmetry}")
    
    # Draw symmetry lines on the image
    draw_symmetry_lines(image_path, vertical_symmetry, horizontal_symmetry)

def detect_symmetry(image, direction='vertical'):
    if direction == 'vertical':
        mid = image.shape[1] // 2
        left_half = image[:, :mid]
        right_half = image[:, mid:]
        
        # Mirror the right half
        mirrored_right_half = cv2.flip(right_half, 1)
        
        # Crop the mirrored part to match the left half size
        mirrored_right_half = mirrored_right_half[:, :left_half.shape[1]]
        
        # Calculate the difference between the left half and the mirrored right half
        difference = cv2.absdiff(left_half, mirrored_right_half)
        
        # Sum differences along the horizontal axis and find the minimum
        min_diff_idx = np.argmin(np.sum(difference, axis=0))
        return min_diff_idx + mid  # Adjust index to original image coordinates

    elif direction == 'horizontal':
        mid = image.shape[0] // 2
        top_half = image[:mid, :]
        bottom_half = image[mid:, :]
        
        # Mirror the bottom half
        mirrored_bottom_half = cv2.flip(bottom_half, 0)
        
        # Crop the mirrored part to match the top half size
        mirrored_bottom_half = mirrored_bottom_half[:top_half.shape[0], :]
        
        # Calculate the difference between the top half and the mirrored bottom half
        difference = cv2.absdiff(top_half, mirrored_bottom_half)
        
        # Sum differences along the vertical axis and find the minimum
        min_diff_idx = np.argmin(np.sum(difference, axis=1))
        return min_diff_idx + mid  # Adjust index to original image coordinates

def draw_symmetry_lines(image_path, vertical_line, horizontal_line):
    img = cv2.imread(image_path)
    img_with_lines = img.copy()
    
    # Draw vertical symmetry line
    cv2.line(img_with_lines, (vertical_line, 0), (vertical_line, img.shape[0]), (0, 255, 0), 2)
    
    # Draw horizontal symmetry line
    cv2.line(img_with_lines, (0, horizontal_line), (img.shape[1], horizontal_line), (0, 255, 0), 2)
    
    # Save and display the result
    cv2.imwrite('symmetry_lines.png', img_with_lines)
    cv2.imshow('Symmetry Lines', img_with_lines)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
find_symmetry_lines('circle12.jpg')
```
